```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/benchmarkVecMapDelete.csv')
```

```python
df
```

```python
ax = None
for container in df['Container'].unique():
    subset = df[df['Container'] == container]
    ax = subset.plot(x='TimeNs', y='Sz', label=container, title='Sz vs Time for each Container', ax=ax)

plt.legend()
plt.show()
```
